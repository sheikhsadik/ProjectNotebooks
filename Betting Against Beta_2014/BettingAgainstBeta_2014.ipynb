**Pulling data from CRSP can be quite tricky, I simply modified a standard CRSP query from online**

List of Variables:

- `permno`: The de facto key to track any stock
- `permco`: The de facto key for any company, the difference between `permno` and `permco` being that a single firm will always have one `permco`, however, they can have multiple share classes, thus can have multiple `permno`
- `date`: Reporting date
- `shrcd`: share code, we are only interested in share codes 10 and 11
- `exchcd`: exchange code, we are interested in 1 and 3
- `ret`: Stock returns
- `retx`: Returns ex dividends
- `shrout`: Shares outstanding
- `prc`: Price

The idea is to merge the monthly CRSP data `msf` with `msenames` to extract the sharecodes and exchange codes using left join. Also the share and exchange codes we are interested in are the ones where the reporting date is between `namedt` and `nameenddt`

```python
# crsp security price data
crsp_query =  """
              select a.permno, a.permco, a.date, b.shrcd, b.exchcd,
              a.ret, a.retx, a.shrout, a.prc, a.cfacshr
              from crsp.msf as a
              left join crsp.msenames as b
              on a.permno=b.permno
              and b.namedt<=a.date
              and a.date<=b.nameendt
              where a.date between '01/01/1920' and '12/31/2020'
              and b.exchcd between 1 and 3
              """

# crsp_df = db.raw_sql(crsp_query)
# crsp_df.to_pickle('crsp_df.pkl')
crsp_df = pd.read_pickle('crsp_df.pkl')

crsp_df['date_me'] = crsp_df['date'] + MonthEnd(0)
crsp_df[['permno', 'permco', 'shrcd', 'exchcd']] = crsp_df[['permno', 'permco', 'shrcd', 'exchcd']].astype(int)
shrcd_ix = (crsp_df['shrcd'] == 10) | (crsp_df['shrcd'] == 11)
crsp_df = crsp_df[shrcd_ix] 
```

**Dealing with delisted securities**

Adjust the CRSP return by incorporating the delisted period return. Delisted return is the return of security after it is delisted.
- It is calculated by comparing a value after delisting against the price on the security’s last trading date.
- The value after delisting can include a price on another exchange or the total value of distributions to shareholders. 
- If there is no opportunity to trade a stock after delisting before it is declared worthless, the value after delisting is zero. 
- Delisting Returns are calculated similarly to total returns except that the value after delisting is used as the current price.

**Negative Prices**

Sometimes you see negative stock prices in CRSP. This means that there was no closing price available for that period. Instead, the bid/ask average was used. To distinguish the bid/ask averages from actual closing prices, CRSP puts a leading dash in front of the price when the bid/ask average was used. If neither price nor bid/ask average is available, Price or Bid/Ask Average is set to zero.

```python
delist_query = """
               select permno, dlret, dlstdt
               from crsp.msedelist
               """

# delist_df = db.raw_sql(delist_query)
# delist_df.to_pickle('delist_df.pkl')
delist_df = pd.read_pickle('delist_df.pkl')

delist_df['permno'] = delist_df['permno'].astype(int)
delist_df['dlstdt'] = pd.to_datetime(delist_df['dlstdt'])
delist_df['date_me']  = delist_df['dlstdt'] + MonthEnd(0)

crsp_df_ = crsp_df.merge(delist_df, how='left', on=['permno', 'date_me'])

crsp_df_['ret'].fillna(0,inplace=True)
crsp_df_['retx'].fillna(0,inplace=True)
crsp_df_['dlret'].fillna(0,inplace=True)
crsp_df_['ret_final'] = (1 + crsp_df_['ret'])*(1 + crsp_df_['dlret']) - 1
crsp_df_['prc'] =  crsp_df_['prc'].abs()
crsp_df_['mcap'] = crsp_df_['prc']*crsp_df_['shrout']
crsp_df_.drop(['dlret','dlstdt','prc'], axis=1, inplace=True)
```

PERMCO is a unique company level identifier that remains unchanged throughout the whole term of company’s existence even if the company changed name/s. PERMNO is a unique stock (share class) level identifier. While most of the companies have one class shares, some companies have more than one class shares traded at different prices, and this is the reason why a company can have more than one PERMNOs.

**Getting the right PERMNOS**

- For each data point, as in for each month and each PERMCO, calculate the maximum market cap availble
- Inner join the maximum market cap vector back to the original dataframe to get the relevant PERMNO for the given PERMCO
- Lastly, replace the max market cap vector with the sum of the market cap for each PERMCO at each date to get the appropriate market cap

```python
mcap_sum = crsp_df_.groupby(['date_me','permco'])['mcap'].sum().reset_index()
mcap_max =  crsp_df_.groupby(['date_me','permco'])['mcap'].max().reset_index()
crsp_df1_ = pd.merge(crsp_df_, mcap_max, how='inner', on=['date_me','permco', 'mcap'])
crsp_df1_.drop('mcap', axis=1, inplace=True)
crsp_df__ = crsp_df1_.merge(mcap_sum, how='inner', on=['date_me','permco']).\
                        sort_values(by=['permno','date_me']).drop_duplicates()
```

**Estimating ex-ante betas** 
- The estimated beta for a given security is given by, 
$$\beta^{TS}_{i,t}= \rho_{i,m,t} \frac{\sigma_{i,t}}{\sigma_{m,t}}$$
- We use a one year rolling standard deviation for the volatilities and a five year horizon for the correlation to account for the fact that correlation appears to move more slowly than volatilities. 
- We require at least six months of non-missing data to estimate volatilities and at least three years for correlation. 
- To reduce the influence of outliers, we shrink the betas, 
$$\beta_{i,t}= w_{i}\beta^{TS}_{i,t} + (1-w_{i})\beta^{XS}_{t}$$
We set $w$ to be 0.6 and $\beta^{XS}$ to be 1.0.

Let's initialize the parameter and get the necessary data ready, 


```python
vol_lkbk = 252
minvol_lkbk = 120
corr_lkbk = 252*5
mincorr_lkbk = 252*3
alpha = 0.6
beta_xs = 1.0

permno_list = crsp_df__['permno'].unique().tolist()
ff_data = data.DataReader('F-F_Research_Data_Factors_daily', 'famafrench', start='1920-10-01')[0][['Mkt-RF','RF']]/100
ff_data.rename(columns= {'Mkt-RF':'mkt_rf','RF':'rf'}, inplace=True)
ff_data.index =  pd.to_datetime(ff_data.index.astype(str))
ff_data['mkt_rf'] = np.log(1+ff_data['mkt_rf'])
ff_data['mkt_rf_vol'] = ff_data['mkt_rf'].rolling(vol_lkbk, min_periods=minvol_lkbk).std()
```

**Estimating betas using CRSP**

Since betas are estimated on daily data, it's faster to send CRSP query for each stock rather than asking for the entire set of permnos, so I basically loop through the unique list of permnos and calculate vector of daily betas, then resample to get a beta estimate for end of each month. 

```python
beta_dict = {}
misc_dict = {}
N = len(permno_list)

for n, permno in enumerate(permno_list):
    
    stock_query = """
                  select a.permno, a.permco, a.date, b.shrcd, b.exchcd, a.ret
                  from crsp.dsf as a
                  left join crsp.dsenames as b
                  on a.permno=b.permno
                  and b.namedt<=a.date
                  and a.date<=b.nameendt
                  where a.date between '01/01/1920' and '12/31/2020'
                  and b.exchcd between 1 and 3
                  and a.permno = {} 
                  """.format(permno)
    stock_df = db.raw_sql(stock_query)
    
    stock_df.index = pd.to_datetime(stock_df['date'])
    stock_df.drop(columns='date', inplace=True)
    
    if len(stock_df.dropna()) < mincorr_lkbk: continue
        
    df = pd.concat([stock_df, ff_data.dropna()], axis=1).dropna()
    df['rx'] = df['ret'] - df['rf']
    df['rx_vol'] = df['rx'].rolling(vol_lkbk, min_periods=minvol_lkbk).std()
    df['rho'] = df[['rx','mkt_rf']].rolling(corr_lkbk, min_periods=mincorr_lkbk).corr(df['rx'])['mkt_rf']
    df['beta'] = df['rho'].multiply(df['rx_vol'].divide(df['mkt_rf_vol'], axis=0), axis=0)
    df['beta'] = df['beta']*alpha + (1-alpha)*beta_xs
    
    beta_dict[permno] = df['beta'].dropna()
    misc_dict[permno] = df[['mkt_rf_vol', 'rx_vol', 'rho']].dropna()
    
    if n % 500 is 0:
        print('{:.2f}% of permnos completed...'.format(n*100/N))

monthly_beta_dict = {}
for permno in list(beta_dict.keys()):
    monthly_beta_dict[permno] = beta_dict[permno].resample('M').last()
    
    
# with open('beta_bab_dict.pkl', 'wb') as handle:
#     pickle.dump(beta_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)
    
# with open('misc_bab_dict.pkl', 'wb') as handle:
#     pickle.dump(misc_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)
    
# with open('monthly_beta_bab_dict.pkl', 'wb') as handle:
#     pickle.dump(monthly_beta_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

**Next, we calculate monthly excess returns for each stock in our universe, we use the fama french risk free rate**

```python
rf_data = data.DataReader('F-F_Research_Data_Factors', 'famafrench', start='1920-10-01')[0][['RF']]/100
rf_data.index =  pd.to_datetime(rf_data.index.astype(str))
rf_data.index = rf_data.index + MonthEnd(-1)
rf_data['date_me'] = rf_data.index
crsp_df__ = crsp_df__.merge(rf_data, how='left', on='date_me')
crsp_df__['rx_final'] = crsp_df__['ret_final'] - crsp_df__['RF']
```

**Portfolio Construction**
- For each time $t$, let $Z$ be the $n x 1$ vector with beta ranks for each stock, 
$$z_{i}= rank(\beta_{i})$$
- The portfolio weight vector for low and high beta portfolios can be calculated as, 

$$W_{H} = k*(Z - \bar{Z})^{+}$$
$$W_{L} = k*(Z - \bar{Z})^{-}$$

Here, $\bar{Z}$ is a $n x 1$ vector with the average rank across all rows. 
$$\bar{Z} = 1^{T}_{n}Z * \frac{1}{n} * 1_{n}$$ 
And, $k$ can be written as, 
$$k = 2/1^{T}_{n}|Z - \bar{Z}|$$


```python
beta_df = pd.concat(monthly_beta_dict, axis=0, keys=monthly_beta_dict.keys()).reset_index()\
                    .rename(columns={'level_0': 'permno','level_1': 'date_me'})

crsp_df__ = crsp_df__.merge(beta_df, how='left', on=['permno', 'date_me'])
crsp_df__['beta_lag'] = crsp_df__.groupby('permno')['beta'].shift(1)
crsp_df__['beta_rank'] = crsp_df__.groupby('date_me')['beta_lag'].rank()
crsp_df__['w_beta'] = crsp_df__.groupby('date_me')['beta_rank'].\
                        apply(lambda x: (2/(x - x.mean()).abs().sum())*(x - x.mean()))
crsp_df__ = crsp_df__[~crsp_df__['w_beta'].isnull()]

crsp_df__['beta_rank'] = crsp_df__['w_beta'].copy() 
crsp_df__['beta_rank'] = np.where(crsp_df__['w_beta'] >= 0.0, 'H', crsp_df__['beta_rank'])
crsp_df__['beta_rank'] = np.where(crsp_df__['w_beta'] < 0.0, 'L', crsp_df__['beta_rank'])
crsp_df__['w_beta'] = crsp_df__['w_beta'].abs()
crsp_df__['decile_rank'] = crsp_df__.groupby('date_me')['beta_lag'].apply(pd.qcut, 10, list(range(1, 11)))
```

**Next, we compute 10 decile portfolios sorted by their betas**

```python
bab_deciles = crsp_df__.groupby(['date_me', 'decile_rank'])['rx_final'].mean().\
                        reset_index().pivot_table(values='rx_final',
                                                  index = 'date_me',
                                                  columns='decile_rank')

bab_deciles.cumsum().plot()
plt.title('Beta sorted decile portfolios')
```

**Calculating the Betting Against Beta factor**
$$r^{BAB}_{t+1} =  \frac{1}{\beta_{L}}*(r^{L}_{t+1}-r_{f}) - \frac{1}{\beta_{H}}*(r^{H}_{t+1}-r_{f})$$
$$\beta_{L}= \beta^{T}W_{L}$$
$$\beta_{H}= \beta^{T}W_{H}$$
$$r^{L}= r^{T}W_{L}$$
$$r^{H}= r^{T}W_{H}$$

```python
babret_df = crsp_df__.groupby(['date_me','beta_rank'])['w_beta', 'rx_final'].\
                apply(lambda x: (x['w_beta']*x['rx_final']).sum()).reset_index().rename(columns={0: 'rx_final'})

babbeta_df = crsp_df__.groupby(['date_me','beta_rank'])['w_beta', 'beta_lag'].\
                apply(lambda x: (x['w_beta']*x['beta_lag']).sum()).reset_index().rename(columns={0: 'beta'})

bab_df = babret_df.merge(babbeta_df, on=['date_me', 'beta_rank'], how='left')
bab_pivot_df = bab_df.pivot_table(values = ['beta','rx_final'],
                                  index = 'date_me',
                                  columns='beta_rank').droplevel(axis=1, level=0)
bab_pivot_df.columns = ['beta_H', 'beta_L', 'rx_H', 'rx_L']
bab_pivot_df['BAB'] = ((bab_pivot_df['beta_L']**(-1))*bab_pivot_df['rx_L']) -  ((bab_pivot_df['beta_H']**(-1))*bab_pivot_df['rx_H'])
bab_pivot_df[['B' + str(pix) for pix in list(range(1, 11))]] = bab_deciles.copy()
```

**Calculate performance stats for the decile portfolios and the Factor**

```python
ff_data = data.DataReader('F-F_Research_Data_Factors_daily', 'famafrench', start='1920-10-01')[0]/100
ff_data = ff_data.apply(lambda x: (1+x)).cumprod().resample('M').last().pct_change(1).dropna()
# ff_data.index =  pd.to_datetime(ff_data.index.astype(str))
# ff_data.index = ff_data.index + MonthEnd(-1)
bab_pivot_df['MKT'] = ff_data['Mkt-RF'].copy()
bab_pivot_df['AQR_BAB'] = pd.read_excel('AQR_bab.xlsx', index_col=0)
```

```python
window_start = '1950/01'
window_fin = '2012/12'

def get_strat_stats(x):
    excess_mu = x.mean()
    excess_vol = x.std()*np.sqrt(12)
    sharpe = (excess_mu*12)/excess_vol
    beta_expost = sm.OLS(x, sm.add_constant(bab_pivot_df.loc[window_start:window_fin, 'MKT']), missing='drop').fit().params[1]
    return pd.Series([excess_mu, excess_vol, sharpe, beta_expost], index=['Monthly Excess Return', 'Volatility', 'Sharpe', 'Ex-post Beta'])
    
bab_pivot_df.loc[window_start:window_fin,'BAB':'B10'].apply(get_strat_stats)
```
